# Text Classification Notebook

This notebook works on text classification using raw text and labels.
It begins by loading the dataset and reviewing the text samples.
The workflow then covers cleaning, tokenization, and feature creation.
Later cells support training, testing, and evaluating classification models.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
import nltk
```

```python
data = {

  "text": [

    "I love working with data and statistics",

    "Data visualization helps understand trends",

    "Data analysis improves business decisions",

    "Machine learning models can predict outcomes",

    "Training models requires a lot of data",

    "Supervised learning uses labeled data",

    "AI is transforming industries rapidly",

    "Artificial intelligence powers smart assistants",

    "AI systems can automate complex tasks",

    "Machine learning improves recommendation systems",

    "Data science combines statistics and programming",

    "Artificial intelligence enables self driving cars"

  ],

   

  "label": [

    "data_science",

    "data_science",

    "data_science",

    "machine_learning",

    "machine_learning",

    "machine_learning",

    "artificial_intelligence",

    "artificial_intelligence",

    "artificial_intelligence",

    "machine_learning",

    "data_science",

    "artificial_intelligence"

  ]

}

df = pd.DataFrame(data)
df
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
df["clean_text"] = df["text"].str.lower()
df
```

```python
df["clean_text"]=df["clean_text"].str.replace(f" [{string.punctuation}]", "", regex=True)
df.head()
```

```python
df["tokens"]=df["clean_text"].apply(word_tokenize)
df
```

```python
stop_words = set(stopwords.words("english"))
df["tokens"] = df["tokens"].apply(lambda x: [word for word in x if word not in stop_words])
df
```

```python
lemmatizer = WordNetLemmatizer()
df["lemma"]=df["tokens"].apply(lambda words:[lemmatizer.lemmatize(w) for w in words])
df
```

```python
df["preprocessed_text"] = df["lemma"].apply(lambda x: " ".join(x))
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_features = vectorizer.fit_transform(df["preprocessed_text"])
tfidf_df = pd.DataFrame(tfidf_features.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df
```

```python
x=tfidf_df
y=df["label"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
accuracy = accuracy_score(y, y_pred)
accuracy
```

```python
#New Prediction 
new_text = ["AI helps automate tasks"]
```

```python
new_text_tfidf = vectorizer.transform(new_text)
new_text_df = pd.DataFrame(new_text_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
prediction = model.predict(new_text_df)
print(prediction)
```
